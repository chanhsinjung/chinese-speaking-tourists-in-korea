# 형태소 분석

```
#thulac(user_dict=None, model_path=None, T2S=False, seg_only=False, filt=False)
#thulac 모델 실행하기
thu1 = thulac.thulac()
```

```
# 사용자 사전
thu1 = thulac.thulac(user_dict='Analysis_사용자사전.txt')
```

```
# "원본데이터"를 대상으로 "thulac"를 사용해서 형태소 분석하기
저장공간  = [] ## 모든 결과를 저장할 빈 공간으로 "저장공간" 설정
for index, row in 원본데이터.iterrows():# 원본데이터의 컬럼(열)을 불러와서 반복할 준비!
    textdata = row[7]  # 분석에서 사용할 텍스트 정보가 있는 열을 지정해준다. 주의! python은 0부터 시작한다!
    type1 = row[15]  # 분석에서 사용할 분류 정보가 있는 열을 지정해준다. 주의! python은 0부터 시작한다!
    type2 = row[2]
    형태소 = thu1.cut(textdata)  # thulac의 분절 기능을 활용해서 대상 텍스트를 형태소 분리한다.
    for word,tag in 형태소:  # thulac로 도출된 개별 형태소에 추가 정보를 기입하기 위해서 반복해준다.
      형태소종합 = [word,tag,type1,type2,1]  # 형태소종합에 분류정보와 카운트정보를 추가한다.
      저장공간.append(형태소종합) # 형태소종합의 내용을 종합하여 저장한다.
```

```
# "형태소" 변수의 저장 내용 확인
형태소
```

```
# "형태소종합" 변수의 저장 내용 확인
형태소종합
```

```
# "저장공간" 변수의 저장 내용 확인
저장공간
```

```
# 단일 데이터프레임 구조로 변환
분석통합 = pd.DataFrame(저장공간)
```

```
# "분석통합" 변수의 저장 내용 확인
분석통합
```

```
# "분석통합" 변수의 컬럼(열) 이름 변경
분석통합.columns = ["형태소", "품사", "지역", "관광지", "카운트"]
```

```
# "분석통합" 변수의 저장 내용 확인
분석통합
```

```
# "분석통합" 변수의 내용을 형태소, 품사, 분류가 같은 것을 합치고, 카운트의 총합을 구해서 "그룹통합" 변수에 저장한다.
그룹통합 = 분석통합.groupby(['형태소', '품사', "지역", "관광지"])['카운트'].sum()
```

```
# "그룹통합" 변수의 저장 내용 확인
그룹통합
```

```
# "그룹통합" 변수의 내용을 데이터프레임 형식으로 변환해서 "그룹통합" 변수에 저장한다.
그룹통합 = pd.DataFrame(그룹통합)
```

```
# "그룹통합" 변수의 저장 내용 확인
그룹통합
```

```
# "그룹통합" 변수의 내용을 "Result_THULAC.csv" 파일로 저장한다.
# header는 컬럼(열) 정보의 포함 여부이다.
# encoding은 문자코드를 선택하는 것이다. python에서는 기본적으로 utf-8(유니코드)를 사용한다.

그룹통합.to_csv('Result_THULAC.csv', header='true', encoding='utf-8')

# 품사태그 대조표 https://github.com/thunlp/THULAC-Python?tab=readme-ov-file#%E8%AF%8D%E6%80%A7%E8%A7%A3%E9%87%8A
```
